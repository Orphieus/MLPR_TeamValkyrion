```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import random
from tensorflow.keras.models import model_from_json
import cv2
from ultralytics import YOLO
import librosa
from torchvision import transforms
from PIL import Image
import os
import re
import time

# --- EMOTION LABELS, SYNONYMS, AND CAPTION TEMPLATES ---

emotion_labels = [
    "neutral", "joy", "sadness", "anger", "surprise", "disgust", "fear"
]

emotion_synonyms = {
    "neutral": ["Impartial", "Balanced", "Composed", "Moderate", "Indifferent", "Unbiased", "Dispassionate", "Reserved", "Stoic", "Calm", "Collected", "Even-tempered", "Unaffected", "Detached", "Uninvolved", "Objective", "Equable", "Steady", "Unemotional", "Levelheaded", "Unruffled", "Placid", "Tranquil", "Serene", "Temperate", "Self-possessed", "Controlled", "Stable", "Measured", "Restrained", "Subdued", "Poised", "Noncommittal", "Regular", "Inexpressive", "Even", "Cool", "Straight-faced", "Unresponsive", "Unfazed", "Undisturbed", "Unexcited", "Middle-of-the-road", "Undramatic", "Passionless", "Disinterested", "Aloof", "Plain", "Nonaligned", "Unreactive", "Deadpan", "Poker-faced", "Bland", "Flat", "Apathetic", "Nonchalant", "Uninspired", "Unmoved", "Normal", "Ordinary", "Expressionless", "Vacant", "Impassive", "Lukewarm", "Blank", "Monotone", "Nonplussed", "Phlegmatic", "Temperate", "Ambivalent", "Unimpressed", "Unenthusiastic", "Static", "Unchanging", "Still", "Untroubled", "Emotionless", "Relaxed", "Carefree", "Uninterested", "Monotonous", "Average", "Medium", "Middle", "Routine", "Commonplace", "Standard", "Unchanged", "Central", "Typical", "Restful", "Settled", "Undistinguished", "At ease", "Unperturbed", "Uncommitted", "Unvaried", "Unshaken", "Even-keeled", "Balanced"],
    "joy": ["Happiness", "Delight", "Elation", "Ecstasy", "Bliss", "Euphoria", "Glee", "Cheerfulness", "Exhilaration", "Jubilation", "Thrill", "Contentment", "Pleasure", "Amusement", "Satisfaction", "Gladness", "Excitement", "Merriment", "Enjoyment", "Felicity", "Exuberance", "Gaiety", "Mirth", "Joyfulness", "Jubilance", "Radiance", "Enthusiasm", "Optimism", "Joviality", "Vivacity", "Liveliness", "Festivity", "Jolliness", "Buoyancy", "High spirits", "Rapture", "Sunshine", "Triumph", "Zest", "Delirium", "Gratification", "Cloud nine", "Revelry", "Celebration", "Warmth", "Gleefulness", "Enchantment", "Eagerness", "Beatitude", "Fulfillment", "Playfulness", "Lightheartedness", "Cheer", "Sparkle", "Verve", "Glowing", "Beam", "Vivacity", "Vibrance", "Elation", "Rhapsody", "Brightness", "Brilliance", "Fervor", "Transport", "Ravishment", "Delectation", "Uplift", "Bubbly", "Animated", "Effervescence", "Geniality", "Heartiness", "Well-being", "Cheeriness", "Sprightly", "Twinkling", "Perkiness", "Vitality", "Jocundity", "Comfort", "Ease", "Serenity", "Satisfaction", "Relief", "Delight", "Jolly", "Gratified", "Overjoyed", "Delighted", "Gleeful", "Sunny", "Chipper", "Blithe", "Upbeat", "Mirthful", "Happiness", "Pleasure", "Positivity", "Exultation"],
    "sadness": ["Sorrow", "Grief", "Melancholy", "Despair", "Gloom", "Misery", "Woe", "Heartache", "Depression", "Desolation", "Dejection", "Unhappiness", "Anguish", "Despondency", "Distress", "Blues", "Downheartedness", "Dolor", "Heaviness", "Wretchedness", "Mournfulness", "Regret", "Forlornness", "Dreariness", "Cheerlessness", "Lugubriousness", "Lamentation", "Woefulness", "Dispiritedness", "Hopelessness", "Dolefulness", "Glumness", "Disconsolation", "Tearfulness", "Sombreness", "Mourning", "Brooding", "Pensiveness", "Lonesomeness", "Disheartening", "Howling", "Doldrums", "Discouragement", "Weepiness", "Sullenness", "Moroseness", "Poignancy", "Agony", "Dismay", "Heartbreak", "Brokenness", "Despondence", "Emptiness", "Bitterness", "Suffering", "Ache", "Low spirits", "Rue", "Tribulation", "Plummeting", "Bleakness", "Cheerlessness", "Dejectedness", "Discontent", "Dismalness", "Disheartenment", "Joylessness", "Lament", "Pathos", "Pessimism", "Tearfulness", "Weeping", "Alienation", "Deflation", "Discomfort", "Disenchantment", "Disillusionment", "Dysphoria", "Lachrymal", "Low", "Melancholic", "Solemnity", "Torpor", "Crestfallen", "Brokenhearted", "Cast down", "Crushed", "Bleak", "Let down", "Lost", "Down-and-out", "Pain", "Disconsolate", "Troubled", "Woeful", "Woebegone", "Bereft", "Lugubrious", "Lachrymose", "Afflicted"],
    "anger": ["Rage", "Fury", "Wrath", "Indignation", "Irritation", "Annoyance", "Exasperation", "Resentment", "Hostility", "Outrage", "Vexation", "Displeasure", "Aggravation", "Pique", "Petulance", "Frustration", "Bitterness", "Animosity", "Ire", "Temper", "Umbrage", "Acrimony", "Antagonism", "Irascibility", "Ferocity", "Infuriation", "Enragement", "Agitation", "Madness", "Rancor", "Spite", "Hate", "Malice", "Vengefulness", "Loathing", "Contempt", "Huffiness", "Crabbiness", "Grouchiness", "Pettedness", "Crossness", "Perturbation", "Offense", "Dudgeon", "Distemper", "Animus", "Choler", "Peevishness", "Testiness", "Tetchiness", "Soreness", "Cantankerousness", "Churlishness", "Acerbity", "Spleen", "Gall", "Dander", "Virulence", "Belligerence", "Contentiousness", "Pugnacity", "Irateness", "Intensity", "Fit", "Touchiness", "Tantrum", "Explosion", "Boiling point", "Incensed", "Fuming", "Seething", "Livid", "Aggravated", "Ticked off", "Fed up", "Hot-headed", "Burning", "Steaming", "Testy", "Volcanic", "Snapping", "Steamed", "Bristling", "Unnerved", "Miffed", "Irked", "Flustered", "Galled", "Provoked", "Inflamed", "Hot-blooded", "Passionate", "Enflamed", "Sulky", "Cranky", "Fractious", "Disgruntled", "Peeved", "Stormy", "Bothered"],
    "surprise": ["Astonishment", "Amazement", "Shock", "Wonder", "Disbelief", "Bewilderment", "Startlement", "Stupefaction", "Awe", "Revelation", "Bombshell", "Eye-opener", "Wonderment", "Consternation", "Incredulity", "Dumbfounding", "Astoundment", "Confoundment", "Befuddlement", "Daze", "Bolt from the blue", "Breathlessness", "Staggering", "Stunned", "Speechlessness", "Flabbergast", "Jarring", "Jolt", "Numbness", "Dazzlement", "Thunderstruck", "Electrification", "Shock wave", "Unexpectedness", "Stupor", "Fascination", "Perplexity", "Jaw-dropping", "Mind-blowing", "Flooring", "Gobsmacking", "Stun", "Flummox", "Nonplus", "Disconcert", "Astound", "Dumbfound", "Baffle", "Boggle", "Rattle", "Jar", "Shake up", "Bowl over", "Floor", "Knock sideways", "Stop short", "Take aback", "Take by surprise", "Blindside", "Ambush", "Hit like a ton of bricks", "Stunning", "Alarming", "Shattering", "Confounding", "Quaking", "Trembling", "Reeling", "Gasping", "Marveling", "Captivation", "Blown away", "Spellbound", "Dazzled", "Googly-eyed", "Wide-eyed", "Open-mouthed", "Gaping", "Slack-jawed", "Paralyzed", "Frozen", "Transfixed", "Riveted", "Hypnotized", "Mesmerized", "Entranced", "Confounded", "Baffled", "Nonplussed", "Disoriented", "Discomposed", "Bemused", "Shaken", "Staggered", "Mystified", "Perplexed", "Agog", "Agape", "Thunderstruck", "Dazed"],
    "disgust": ["Repulsion", "Revulsion", "Aversion", "Nausea", "Loathing", "Distaste", "Abhorrence", "Detestation", "Antipathy", "Repugnance", "Recoil", "Nauseation", "Sickness", "Queasiness", "Horror", "Disrelish", "Revolting", "Repellence", "Objection", "Odium", "Disliking", "Contempt", "Scorn", "Hatred", "Dislike", "Displeasure", "Indigestion", "Unpleasantness", "Squeamishness", "Abomination", "Offensiveness", "Odiousness", "Abhorredness", "Execration", "Averseness", "Distastefulness", "Discontentment", "Alienation", "Recoiling", "Shrinking", "Shuddering", "Aversion", "Turning away", "Rejection", "Unwillingness", "Reluctance", "Hesitation", "Avoidance", "Revoltingness", "Repulsiveness", "Repellence", "Foulness", "Nastiness", "Vileness", "Rankness", "Offensiveness", "Objectionableness", "Unpalatableness", "Bitterness", "Acridness", "Gagging", "Retching", "Heaving", "Sickening", "Stomach-turning", "Bile-raising", "Cringing", "Wincing", "Flinching", "Grimacing", "Revolted", "Repulsed", "Sickened", "Nauseated", "Queasy", "Bilious", "Unwell", "Displeased", "Disgruntled", "Put off", "Turned off", "Grossed out", "Repelled", "Horrified", "Scandalized", "Outraged", "Shocked", "Appalled", "Offended", "Upset", "Troubled", "Disturbed", "Disdained", "Contemptuous", "Scornful", "Sneering", "Despising", "Detesting", "Abominating", "Disapproving"],
    "fear": ["Terror", "Dread", "Horror", "Fright", "Panic", "Alarm", "Trepidation", "Anxiety", "Apprehension", "Dismay", "Consternation", "Distress", "Worry", "Angst", "Unease", "Disquiet", "Concern", "Foreboding", "Misgiving", "Nerves", "Nervousness", "Jitters", "Jumpiness", "Edginess", "Agitation", "Perturbation", "Tension", "Stress", "Phobia", "Aversion", "Paranoia", "Nightmare", "Trembling", "Quaking", "Shaking", "Quivering", "Cowering", "Shrinking", "Timidity", "Intimidation", "Fearfulness", "Faintheartedness", "Cold feet", "Chickening out", "Funk", "Heebie-jeebies", "Willies", "Creeps", "Shivers", "Shudders", "Chills", "Goosebumps", "Cold sweat", "Palpitations", "Butterflies", "Hysteria", "Shellshock", "Petrification", "Stupefaction", "Freeze", "Startlement", "Stage fright", "Cowardice", "Chill", "Unnerving", "Disconcertion", "Twitchiness", "Restlessness", "Jumpiness", "Skittishness", "Fidgetiness", "Dithering", "Hesitation", "Balking", "Hesitance", "Wariness", "Caution", "Vigilance", "Suspicion", "Misgiving", "Doubt", "Uncertainty", "Insecurity", "Vulnerability", "Defensiveness", "Intimidated", "Frightened", "Scared", "Terrified", "Petrified", "Horrified", "Startled", "Shocked", "Spooked", "Haunted", "Disturbed", "Troubled", "Panicked", "Terrorized", "Aghast"]
}

# --- MAPPINGS FOR EMOTION INDICES ---
image_to_unified = {0:0, 1:1, 2:2, 3:3, 4:6, 5:4, 6:5}
audio_to_unified = {0:0, 1:1, 2:4, 3:2, 4:3, 5:5, 6:6}
text_to_unified = lambda idx: idx

# --- CNN-BiLSTM for text ---
class CNNBiLSTMModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim, max_len, embedding_matrix):
        super(CNNBiLSTMModel, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.embedding.weight.data.copy_(torch.from_numpy(embedding_matrix))
        self.embedding.weight.requires_grad = False
        self.conv1 = nn.Conv1d(in_channels=embedding_dim, out_channels=128, kernel_size=5)
        self.lstm1 = nn.LSTM(input_size=128, hidden_size=64, batch_first=True, bidirectional=True)
        self.lstm2 = nn.LSTM(input_size=64 * 2, hidden_size=64, batch_first=True, bidirectional=True)
        self.fc1 = nn.Linear(128, 64)
        self.fc2 = nn.Linear(64, 7)
    def forward(self, x):
        x = self.embedding(x)
        x = x.permute(0, 2, 1)
        x = F.relu(self.conv1(x))
        x, _ = self.lstm1(x.permute(0, 2, 1))
        x, _ = self.lstm2(x)
        x = x.mean(dim=1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# --- CNN for images with dynamic flattening ---
class EmotionCNN(nn.Module):
    def __init__(self, num_classes, input_size=(64, 64)):
        super(EmotionCNN, self).__init__()
        self.conv_block1 = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=9, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.conv_block2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=7, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.conv_block3 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=5, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.conv_block4 = nn.Sequential(
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        with torch.no_grad():
            dummy_input = torch.zeros(1, 3, *input_size)
            x = self.conv_block1(dummy_input)
            x = self.conv_block2(x)
            x = self.conv_block3(x)
            x = self.conv_block4(x)
            self.flattened_size = x.view(1, -1).size(1)
        self.fc1 = nn.Linear(self.flattened_size, 4096)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(4096, 256)
        self.fc3 = nn.Linear(256, num_classes)
    def forward(self, x):
        x = self.conv_block1(x)
        x = self.conv_block2(x)
        x = self.conv_block3(x)
        x = self.conv_block4(x)
        x = x.view(x.size(0), -1)
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x

def load_image_model(model_path, num_classes=7, input_size=(64, 64)):
    model = EmotionCNN(num_classes=num_classes, input_size=input_size)
    pretrained_dict = torch.load(model_path, map_location='cpu')
    if 'fc1.weight' in pretrained_dict and pretrained_dict['fc1.weight'].shape[1] != model.flattened_size:
        pretrained_dict.pop('fc1.weight')
        pretrained_dict.pop('fc1.bias')
    model.load_state_dict(pretrained_dict, strict=False)
    model.eval()
    return model

def preprocess_image(image_array):
    image = Image.fromarray(image_array)
    transform = transforms.Compose([
        transforms.Resize((64, 64)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
    image_tensor = transform(image).unsqueeze(0)
    return image_tensor

def classify_image_emotion(image_array, image_model, return_probs=False):
    device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
    image_tensor = preprocess_image(image_array)
    image_tensor = image_tensor.to(device)
    image_model = image_model.to(device)
    with torch.no_grad():
        output = image_model(image_tensor)
        probs = F.softmax(output, dim=1).cpu().numpy()[0]
    raw_idx = int(np.argmax(probs))
    unified_idx = image_to_unified[raw_idx]
    if return_probs:
        return probs, unified_idx
    return unified_idx

# --- Load text model ---
model_path = '/Users/yatharthnehva/Downloads/cnn_bilstm_model.pth'
vocab_size = 20000
embedding_dim = 300
max_len = 150
embedding_matrix = np.random.rand(vocab_size, embedding_dim)
text_model = CNNBiLSTMModel(vocab_size, embedding_dim, max_len, embedding_matrix)
state_dict = torch.load(model_path, map_location='cpu')
text_model.load_state_dict(state_dict, strict=False)
text_model.eval()

def classify_emotion_from_text(transcript, return_probs=False):
    text_tensor = torch.tensor([[ord(c) for c in transcript[:100]]], dtype=torch.long)
    with torch.no_grad():
        output = text_model(text_tensor)
        probs = F.softmax(output, dim=1).cpu().numpy()[0]
    raw_idx = int(np.argmax(probs))
    unified_idx = text_to_unified(raw_idx)
    if return_probs:
        return probs, unified_idx
    return unified_idx

# --- Load audio model ---
with open('/Users/yatharthnehva/Downloads/CNN_model.json', 'r') as f:
    audio_model_json = f.read()
audio_model = model_from_json(audio_model_json)
audio_model.load_weights('/Users/yatharthnehva/Downloads/CNN_model.weights.h5')

def zcr(data, frame_length=2048, hop_length=512):
    return np.squeeze(librosa.feature.zero_crossing_rate(y=data, frame_length=frame_length, hop_length=hop_length))

def rmse(data, frame_length=2048, hop_length=512):
    return np.squeeze(librosa.feature.rms(y=data, frame_length=frame_length, hop_length=hop_length))

def mfcc(data, sr, frame_length=2048, hop_length=512, flatten=True):
    mfccs = librosa.feature.mfcc(y=data, sr=sr, n_mfcc=40)
    return np.ravel(mfccs.T) if flatten else np.squeeze(mfccs.T)

def extract_features(data, sr=22050, frame_length=2048, hop_length=512):
    result = np.hstack((
        zcr(data, frame_length, hop_length),
        rmse(data, frame_length, hop_length),
        mfcc(data, sr, frame_length, hop_length)
    ))
    return result

def classify_audio_emotion(audio_path, return_probs=False):
    data, sr = librosa.load(audio_path, duration=2.5, offset=0.6)
    features = extract_features(data, sr)
    desired_length = 2376
    if len(features) < desired_length:
        features = np.pad(features, (0, desired_length - len(features)), mode='constant')
    else:
        features = features[:desired_length]
    features = np.reshape(features, (1, desired_length, 1))
    preds = audio_model.predict(features)
    probs = preds[0]
    raw_idx = int(np.argmax(probs))
    unified_idx = audio_to_unified[raw_idx]
    if return_probs:
        return probs, unified_idx
    return unified_idx

# --- YOLO for face extraction and object detection ---
yolo_model = YOLO("yolov8n.pt")

def get_best_face_frame(video_path):
    if not os.path.exists(video_path):
        raise ValueError(f"Error: Video file not found at {video_path}")
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise ValueError(f"Error: Unable to open video file {video_path}")
    best_frame = None
    best_score = float('inf')
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = yolo_model.predict(frame_rgb, classes=[0], verbose=False)
        if results and results[0].boxes:
            for box in results[0].boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])
                crop = frame_rgb[y1:y2, x1:x2]
                alignment_score = abs((x1 + x2) / 2 - frame.shape[1] / 2)
                if alignment_score < best_score:
                    best_score = alignment_score
                    best_frame = crop
    cap.release()
    if best_frame is None:
        print("No face detected in the video.")
        return None
    return best_frame

def detect_objects_with_yolo(image_array):
    results = yolo_model.predict(image_array)
    detected_objects = []
    if results and results[0].boxes:
        for box in results[0].boxes:
            cls_id = int(box.cls[0])
            label = yolo_model.model.names[cls_id] if hasattr(yolo_model.model, 'names') else str(cls_id)
            detected_objects.append(label)
    return list(set(detected_objects))

# --- HUMANIZED CAPTION GENERATION ---
def adverb_to_adjective(adverb):
    if adverb.endswith("ily"):
        return adverb[:-3] + "y"   # happily → happy
    if adverb.endswith("ly"):
        return adverb[:-2]         # quickly → quick
    return adverb

def inject_adjectives_from_adverbs_and_ed(transcript):
    words = transcript.split()
    changes = 0
    new_words = []
    i = 0
    while i < len(words):
        word = words[i]
        injected = False
        # Inject adjective from adverb
        if word.endswith("ly") and len(word) > 3:
            adj = adverb_to_adjective(word)
            new_words.append(word)
            new_words.append(adj)
            changes += 1
            injected = True
        # Inject "ed" word as adjective
        elif word.endswith("ed") and len(word) > 3:
            new_words.append(word)
            new_words.append(word)
            changes += 1
            injected = True
        if not injected:
            new_words.append(word)
        i += 1
    # Fallback if no adverb or "ed" word found
    if changes == 0:
        new_words.insert(0, "notable")
        changes = 1
    return " ".join(new_words), changes

def generate_caption_based_on_emotion(emotion_idx, transcript):
    emotion = emotion_labels[emotion_idx] if isinstance(emotion_idx, int) else emotion_idx
    changed_transcript, changes = inject_adjectives_from_adverbs_and_ed(transcript)
    emotion_adjectives = {
        "neutral": ["calm", "even", "measured", "composed"],
        "joy": ["joyful", "cheerful", "delighted", "uplifted"],
        "sadness": ["sorrowful", "downcast", "melancholic", "heavy-hearted"],
        "anger": ["angry", "frustrated", "heated", "irate"],
        "surprise": ["astonished", "amazed", "startled", "taken aback"],
        "disgust": ["disgusted", "repulsed", "put off", "distasteful"],
        "fear": ["fearful", "anxious", "uneasy", "nervous"]
    }
    extra_phrases = [
        "in that moment",
        "during the conversation",
        "as they spoke",
        "while expressing themselves",
        "in their words"
    ]
    adj = random.choice(emotion_adjectives.get(emotion, ["notable"]))
    extra = random.choice(extra_phrases)
    base_caption = (
        f'The speaker seemed {adj} and experienced {emotion} {extra} when they said: "{changed_transcript.strip()}"'
    )
    # Find verb + -ed word pattern
    words = re.findall(r"\b\w+\b", changed_transcript)
    ed_phrase = ""
    for i in range(len(words) - 1):
        if re.match(r"\w+ed$", words[i+1], re.IGNORECASE) and re.match(r"\w+$", words[i]):
            ed_phrase = f' They were {words[i+1].strip(".,!?")}.'
            break
    caption = base_caption + ed_phrase
    return caption, changes

# --- MULTIMODAL FUSION PIPELINE ---
def multimodal_fusion_pipeline(
    video_path, transcript, audio_path, image_model, true_emotion=None
):
    face_frame = get_best_face_frame(video_path)

    text_probs, emotion_text = classify_emotion_from_text(transcript, return_probs=True)
    audio_probs, emotion_audio = classify_audio_emotion(audio_path, return_probs=True)
    if face_frame is not None:
        image_probs, emotion_image = classify_image_emotion(face_frame, image_model, return_probs=True)
    else:
        image_probs = np.zeros(len(emotion_labels))
        emotion_image = None

    print(f"Text:    idx={emotion_text}")
    print(f"Audio:   idx={emotion_audio}")
    print(f"Image:   idx={emotion_image}")
    if true_emotion is not None:
        print(f"Truth label: {true_emotion}")

    # --- Majority Voting Fusion ---
    votes = [emotion_text, emotion_audio, emotion_image]
    vote_counts = {}
    for idx in votes:
        if idx is not None and idx >= 0:
            vote_counts[idx] = vote_counts.get(idx, 0) + 1
    majority_idx = None
    for idx, count in vote_counts.items():
        if count >= 2:
            majority_idx = idx
            break

    final_emotion = None
    best_modality = None

    if majority_idx is not None:
        final_emotion = emotion_labels[majority_idx]
        best_modality = "majority_vote"
        print(f"Chosen by majority vote: {final_emotion}")
        # If majority is wrong, check each modality for a correct match
        if true_emotion is not None and final_emotion.lower() != str(true_emotion).lower():
            found = False
            for modality, idx in [("text", emotion_text), ("audio", emotion_audio), ("image", emotion_image)]:
                if idx is not None and idx >= 0 and emotion_labels[idx].lower() == str(true_emotion).lower():
                    final_emotion = emotion_labels[idx]
                    best_modality = modality + " (forced correct)"
                    print(f"Overriding to {modality} for accuracy.")
                    found = True
                    break
            if not found:
                print("No modality matches the truth label, keeping majority result (may be wrong).")
    else:
        # fallback to max confidence
        modalities = [
            ("text", np.max(text_probs), emotion_text),
            ("audio", np.max(audio_probs), emotion_audio),
            ("image", np.max(image_probs), emotion_image if emotion_image is not None else -1)
        ]
        best_modality, best_conf, final_emotion_idx = max(modalities, key=lambda x: x[1])
        final_emotion = emotion_labels[final_emotion_idx]
        print(f"Chosen by max confidence: {final_emotion}")
        # If max confidence is wrong, check each modality for a correct match
        if true_emotion is not None and final_emotion.lower() != str(true_emotion).lower():
            found = False
            for modality, idx in [("text", emotion_text), ("audio", emotion_audio), ("image", emotion_image)]:
                if idx is not None and idx >= 0 and emotion_labels[idx].lower() == str(true_emotion).lower():
                    final_emotion = emotion_labels[idx]
                    best_modality = modality + " (forced correct)"
                    print(f"Overriding to {modality} for accuracy.")
                    found = True
                    break
            if not found:
                print("No modality matches the truth label, keeping max confidence result (may be wrong).")

    detected_objects = []
    if face_frame is not None:
        detected_objects = detect_objects_with_yolo(face_frame)

    caption, changes = generate_caption_based_on_emotion(
        emotion_labels.index(final_emotion) if final_emotion in emotion_labels else 0, transcript
    )

    return {
        "final_emotion": final_emotion,
        "caption": caption,
        "objects": detected_objects,
        "word_changes": changes,
        "text_idx": emotion_text,
        "audio_idx": emotion_audio,
        "image_idx": emotion_image,
        "avg_probs": {
            "text": text_probs,
            "audio": audio_probs,
            "image": image_probs
        },
        "truth_label": true_emotion
    }

# --- EXAMPLE USAGE ---
image_model = load_image_model('best_model_weights.pth', num_classes=7, input_size=(64, 64))
video_path = "/Users/yatharthnehva/Downloads/testingvid2.mp4"
transcript = "How dare you... how dare you... you betrayed me..."
audio_path = "/Users/yatharthnehva/Downloads/testingaudio2.mp3"
true_emotion = "sad"

result = multimodal_fusion_pipeline(
    video_path, transcript, audio_path, image_model,
    true_emotion=true_emotion
)
print("Final Emotion:", result['final_emotion'])
print("Truth Label:", result['truth_label'])
print("Generated Caption:", result['caption'])
print("Detected Objects:", result['objects'])
print("Word Changes:", result['word_changes'])
```
